```python
import os
from groq import Groq
from secret_api_keys import huggingface_api_key
from secret_api_keys import groq_api_key
from huggingface_hub import InferenceClient
os.environ["HUGGINGFACEHUB_API_TOKEN"] = huggingface_api_key
groq_client = Groq(api_key=groq_api_key)
```

```python
# HuggingFace – DeepSeek
deepseek_client = InferenceClient(
    model="deepseek-ai/DeepSeek-V3.2-Exp",
    token=huggingface_api_key
)

# HuggingFace – Meta Llama 3.1
hf_llama_client = InferenceClient(
    model="meta-llama/Meta-Llama-3.1-8B-Instruct",
    token=huggingface_api_key
)

# Groq – Llama 3.1 8B Instant (FREE)
groq_client = Groq(api_key=groq_api_key)

```

```python
def generate_with(model_name, prompt):
    
    # ---- DeepSeek (HuggingFace) ----
    if model_name == "deepseek":
        resp = deepseek_client.chat_completion(
            messages=[{"role": "user", "content": prompt}]
        )
        return resp.choices[0].message["content"]

    # ---- Llama 3.1 (HF) ----
    elif model_name == "hf_llama":
        resp = hf_llama_client.chat_completion(
            messages=[{"role": "user", "content": prompt}]
        )
        return resp.choices[0].message["content"]

    # ---- Llama 3.1 (Groq) ----
    elif model_name == "groq_llama":
        completion = groq_client.chat.completions.create(
            model="llama-3.1-8b-instant",
            messages=[{"role": "user", "content": prompt}]
        )
        return completion.choices[0].message.content

    else:
        raise ValueError("❌ Unknown model name")
```

```python
import sounddevice as sd
from scipy.io.wavfile import write
import soundfile as sf
import numpy as np
import librosa
import torch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
# Load pretrained model + tokenizer
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
def record_audio(duration=5, filename="live.wav"):
    print("🎙️ Recording...")
    sample_rate = 16000
    audio = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=1)
    sd.wait()
    write(filename, sample_rate, audio)
    print("✅ Recording saved as", filename)

def speech_to_text(audio_path):
    speech, sample_rate = librosa.load(audio_path, sr=16000)
    input_values = tokenizer(speech, return_tensors="pt", padding="longest").input_values

    with torch.no_grad():
        logits = model(input_values).logits

    predicted_ids = torch.argmax(logits, dim=-1)
    text = tokenizer.decode(predicted_ids[0])
    return text
```

```python

from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import nltk

def compute_bleu(reference_text, generated_text):
    """
    BLEU Score for similarity between reference text 
    and generated model output.
    """

    # Tokenize (BLEU requires tokenized input)
    reference_tokens = nltk.word_tokenize(reference_text.lower())
    generated_tokens = nltk.word_tokenize(generated_text.lower())

    smoothie = SmoothingFunction().method4

    # BLEU score (0 to 1)
    bleu = sentence_bleu(
        [reference_tokens], 
        generated_tokens, 
        smoothing_function=smoothie
    )

    return bleu


# ================================
# EXAMPLE TEST FOR 3 MODELS
# ================================

reference = """
Machine learning is a method of data analysis that automates analytical model building.
It is a branch of artificial intelligence based on the idea that systems can learn 
from data, identify patterns and make decisions.
"""

groq_output = """
Machine learning is a field of AI where computers learn patterns from data and can
make predictions or decisions without being explicitly programmed.
"""

deepseek_output = """
Machine learning allows systems to learn automatically from past data and improve 
their performance without manual programming.
"""

hf_llama_output = """
Machine learning is an approach where algorithms learn patterns from data to make
_predictions_ and improve performance over time.
"""

print("BLEU Score (Groq):", compute_bleu(reference, groq_output))
print("BLEU Score (DeepSeek):", compute_bleu(reference, deepseek_output))
print("BLEU Score (HF-LLaMA):", compute_bleu(reference, hf_llama_output))
```

```python
import torch
from bert_score import score
from sentence_transformers import SentenceTransformer, util
from rouge_score import rouge_scorer
from transformers import GPT2LMHeadModel, GPT2TokenizerFast
```

```python
# ---------------------------------------
# 1️⃣ BERTScore (Best semantic similarity)
# ---------------------------------------
def compute_bertscore(reference, generated, lang="en"):
    P, R, F1 = score([generated], [reference], lang=lang)
    return float(F1.mean())       # return best metric


# ---------------------------------------
# 2️⃣ Cosine Similarity (Embedding match)
# ---------------------------------------
embedder = SentenceTransformer('all-MiniLM-L6-v2')

def compute_cosine_similarity(reference, generated):
    emb_ref = embedder.encode(reference, convert_to_tensor=True)
    emb_gen = embedder.encode(generated, convert_to_tensor=True)
    cosine_sim = util.pytorch_cos_sim(emb_ref, emb_gen)
    return float(cosine_sim.item())


# ---------------------------------------
# 3️⃣ ROUGE-L (Structure quality)
# ---------------------------------------
rouge = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)

def compute_rouge_l(reference, generated):
    score_output = rouge.score(reference, generated)
    return float(score_output['rougeL'].fmeasure)


# ---------------------------------------
# 4️⃣ Perplexity (Language smoothness)
# ---------------------------------------
tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

def compute_perplexity(text):
    enc = tokenizer(text, return_tensors="pt")
    input_ids = enc.input_ids

    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
    loss = outputs.loss
    ppl = torch.exp(loss)
    return float(ppl)


# ---------------------------------------
# 🔥 MASTER FUNCTION (Use for comparison)
# ---------------------------------------
def evaluate_text(reference, generated):
    return {
        "BERTScore": compute_bertscore(reference, generated),
        "Cosine Similarity": compute_cosine_similarity(reference, generated),
        "ROUGE-L": compute_rouge_l(reference, generated),
        "Perplexity": compute_perplexity(generated)
    }

```

```python
ref = "Machine learning helps computers learn from previous data."

groq = "Machine learning allows computers to study data and improve automatically."
deepseek = "ML is a method where systems improve by learning from past information."
llama = "Machine learning helps systems gain knowledge from data patterns."

print("Groq:", evaluate_text(ref, groq))
print("DeepSeek:", evaluate_text(ref, deepseek))
print("LLaMA:", evaluate_text(ref, llama))
```
